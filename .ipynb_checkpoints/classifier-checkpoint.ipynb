```python
import os, csv, random, numpy as np
from pathlib import Path
from PIL import Image, ImageDraw

random.seed(42)
np.random.seed(42)

IMG_SIZE = 128
TRAIN_PER_CLASS = 500
VAL_PER_CLASS = 100
SHAPES = ["circle", "square", "triangle"]
CONDITIONS = ["FL_FR", "FL_RR", "RL_FR", "RL_RR"]
FIXED_SIZE = 48
FIXED_ROT = 0
SIZE_RANGE = (24, 64)
ROT_RANGE = (0, 360)
RANDOM_BG = True
GAUSS_NOISE = True
NOISE_STD = 12
OUT_DIR = Path("data")

def make_background(img_size, random_bg=True, gauss_noise=True, noise_std=12):
    if random_bg:
        base_color = np.random.randint(0, 256, size=3, dtype=np.uint8)
        bg = np.ones((img_size, img_size, 3), dtype=np.uint8) * base_color
    else:
        bg = np.ones((img_size, img_size, 3), dtype=np.uint8) * 255
    if gauss_noise:
        noise = np.random.normal(0, noise_std, bg.shape).astype(np.int16)
        bg = np.clip(bg.astype(np.int16) + noise, 0, 255).astype(np.uint8)
    return Image.fromarray(bg)

def draw_shape(shape, size, rotation, img_size):
    layer = Image.new("RGBA", (img_size, img_size), (0, 0, 0, 0))
    draw = ImageDraw.Draw(layer)
    cx, cy = img_size // 2, img_size // 2
    half = size // 2
    if shape == "circle":
        bbox = [cx - half, cy - half, cx + half, cy + half]
        draw.ellipse(bbox, fill=(0, 0, 0, 255))
    elif shape == "square":
        bbox = [cx - half, cy - half, cx + half, cy + half]
        draw.rectangle(bbox, fill=(0, 0, 0, 255))
    elif shape == "triangle":
        p1 = (cx, cy - half)
        p2 = (cx - half, cy + half)
        p3 = (cx + half, cy + half)
        draw.polygon([p1, p2, p3], fill=(0, 0, 0, 255))
    else:
        raise ValueError("Unknown shape")
    if rotation != 0:
        layer = layer.rotate(rotation, resample=Image.BICUBIC, expand=False, center=(cx, cy))
    return layer

def gen_one_image(shape, cond):
    size = FIXED_SIZE if "FL" in cond else random.randint(*SIZE_RANGE)
    rot = FIXED_ROT if "FR" in cond else random.uniform(*ROT_RANGE)
    bg = make_background(IMG_SIZE, RANDOM_BG, GAUSS_NOISE, NOISE_STD).convert("RGBA")
    shp_layer = draw_shape(shape, size, rot, IMG_SIZE)
    out = Image.alpha_composite(bg, shp_layer).convert("RGB")
    return out, size, rot

def gen_split(split, n_per_class):
    meta_rows = []
    for cond in CONDITIONS:
        for shape in SHAPES:
            save_dir = OUT_DIR / split / cond / shape
            save_dir.mkdir(parents=True, exist_ok=True)
            for i in range(n_per_class):
                img, size, rot = gen_one_image(shape, cond)
                fname = f"{shape}_{cond}_{i:05d}.png"
                path = save_dir / fname
                img.save(path, "PNG")
                meta_rows.append([str(path), split, cond, shape, size, rot, RANDOM_BG, GAUSS_NOISE])
    return meta_rows

OUT_DIR.mkdir(parents=True, exist_ok=True)
meta = []
meta += gen_split("train", TRAIN_PER_CLASS)
meta += gen_split("val", VAL_PER_CLASS)

csv_path = OUT_DIR / "metadata.csv"
with csv_path.open("w", newline="") as f:
    w = csv.writer(f)
    w.writerow(["path","split","condition","shape","size","rotation","rand_bg","gauss_noise"])
    w.writerows(meta)

print("Done. Samples:", len(meta))
print("Example row:", meta[0])
```

```python
import random, math, numpy as np
from PIL import Image, ImageDraw
random.seed(42)

def make_shape_image(shape,
                     img_size=128,
                     size_range=(24, 64),
                     rot_range=(0, 360),
                     random_bg=True,
                     gauss_noise=False,
                     noise_std=12):
    size = random.randint(*size_range)
    rot  = random.uniform(*rot_range)

    # background
    if random_bg:
        base = np.random.randint(0, 256, size=3, dtype=np.uint8)
        bg = np.ones((img_size, img_size, 3), dtype=np.uint8) * base
    else:
        bg = np.ones((img_size, img_size, 3), dtype=np.uint8) * 255
    if gauss_noise:
        noise = np.random.normal(0, noise_std, bg.shape).astype(np.int16)
        bg = np.clip(bg.astype(np.int16) + noise, 0, 255).astype(np.uint8)
    bg = Image.fromarray(bg).convert("RGBA")

    layer = Image.new("RGBA", (img_size, img_size), (0, 0, 0, 0))
    draw  = ImageDraw.Draw(layer)

    half = size // 2

    # choose a center so the unrotated bbox fits; rotation might extend slightly, but acceptable for small shapes
    cx = random.randint(half, img_size - half)
    cy = random.randint(half, img_size - half)

    if shape == "circle":
        bbox = [cx - half, cy - half, cx + half, cy + half]
        draw.ellipse(bbox, fill=(0, 0, 0, 255))

    elif shape == "square":
        bbox = [cx - half, cy - half, cx + half, cy + half]
        draw.rectangle(bbox, fill=(0, 0, 0, 255))

    elif shape == "triangle":
        p1 = (cx, cy - half)
        p2 = (cx - half, cy + half)
        p3 = (cx + half, cy + half)
        draw.polygon([p1, p2, p3], fill=(0, 0, 0, 255))
    else:
        raise ValueError("shape must be circle/square/triangle")

    if rot != 0:
        layer = layer.rotate(rot, resample=Image.BICUBIC, center=(cx, cy))

    out = Image.alpha_composite(bg, layer).convert("RGB")

    meta = {"shape": shape, "size": size, "rotation": rot, "center": (cx, cy)}
    return out, meta
```

```python
import random, csv, numpy as np
from pathlib import Path
from PIL import Image, ImageDraw

random.seed(42); np.random.seed(42)

IMG_SIZE = 128
N_PER_SHAPE = 600            # total per shape; will be split later
SHAPES = ["circle", "square", "triangle"]
SIZE_RANGE = (24, 64)
ROT_RANGE  = (0, 360)
RANDOM_BG  = True
GAUSS_NOISE= True
NOISE_STD  = 12
ROOT = Path("geom_data_raw")

def make_bg():
    if RANDOM_BG:
        base = np.random.randint(0,256,3, dtype=np.uint8)
        bg = np.ones((IMG_SIZE, IMG_SIZE, 3), dtype=np.uint8)*base
    else:
        bg = np.ones((IMG_SIZE, IMG_SIZE, 3), dtype=np.uint8)*255
    if GAUSS_NOISE:
        noise = np.random.normal(0, NOISE_STD, bg.shape).astype(np.int16)
        bg = np.clip(bg.astype(np.int16)+noise, 0,255).astype(np.uint8)
    return Image.fromarray(bg).convert("RGBA")

def draw_shape(shape, size, rot):
    layer = Image.new("RGBA", (IMG_SIZE, IMG_SIZE), (0,0,0,0))
    d = ImageDraw.Draw(layer)
    half = size//2
    cx = random.randint(half, IMG_SIZE-half)
    cy = random.randint(half, IMG_SIZE-half)
    if shape=="circle":
        d.ellipse([cx-half, cy-half, cx+half, cy+half], fill=(0,0,0,255))
    elif shape=="square":
        d.rectangle([cx-half, cy-half, cx+half, cy+half], fill=(0,0,0,255))
    elif shape=="triangle":
        p1=(cx, cy-half); p2=(cx-half, cy+half); p3=(cx+half, cy+half)
        d.polygon([p1,p2,p3], fill=(0,0,0,255))
    layer = layer.rotate(rot, resample=Image.BICUBIC, center=(cx,cy))
    return layer

def make_one(shape):
    size = random.randint(*SIZE_RANGE)
    rot  = random.uniform(*ROT_RANGE)
    out = Image.alpha_composite(make_bg(), draw_shape(shape,size,rot)).convert("RGB")
    return out, size, rot

ROOT.mkdir(parents=True, exist_ok=True)
rows=[]
for shape in SHAPES:
    for i in range(N_PER_SHAPE):
        img, s, r = make_one(shape)
        path = ROOT / f"{shape}_{i:05d}.png"
        img.save(path)
        rows.append([str(path), shape, s, r])

with open(ROOT/"metadata.csv","w",newline="") as f:
    w=csv.writer(f); w.writerow(["path","shape","size","rotation"]); w.writerows(rows)

len(rows)
```

```python
import pandas as pd, shutil
from sklearn.model_selection import train_test_split

df = pd.read_csv(ROOT/"metadata.csv")

train_df, tmp = train_test_split(df, test_size=0.3, stratify=df["shape"], random_state=42)
val_df, test_df = train_test_split(tmp, test_size=0.5, stratify=tmp["shape"], random_state=42)

def move_subset(sub_df, subset_name):
    for shape in SHAPES:
        out_dir = Path("geom_data")/subset_name/shape
        out_dir.mkdir(parents=True, exist_ok=True)
    for _,row in sub_df.iterrows():
        src = Path(row["path"])
        dst = Path("geom_data")/subset_name/row["shape"]/src.name
        if not dst.exists():
            shutil.copy(src, dst)

move_subset(train_df, "train")
move_subset(val_df,   "val")
move_subset(test_df,  "test")

print(len(train_df), len(val_df), len(test_df))
```

```python
import torch
from torch.utils.data import DataLoader
from torchvision import datasets, transforms

tf_to_tensor = transforms.ToTensor()

train_tmp = datasets.ImageFolder("geom_data/train", transform=tf_to_tensor)
loader_tmp = DataLoader(train_tmp, batch_size=64, shuffle=False, num_workers=2)

mean = 0.; std = 0.; nb = 0
for x, _ in loader_tmp:
    nb += x.size(0)
    mean += x.mean(dim=[0,2,3])*x.size(0)
    std  += x.std(dim=[0,2,3], unbiased=False)*x.size(0)
mean /= nb; std /= nb
print("mean:", mean.tolist(), "std:", std.tolist())
```

```python
tf_train = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean, std)
])
tf_eval = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean, std)
])

train_ds = datasets.ImageFolder("geom_data/train", transform=tf_train)
val_ds   = datasets.ImageFolder("geom_data/val",   transform=tf_eval)
test_ds  = datasets.ImageFolder("geom_data/test",  transform=tf_eval)

BATCH = 64
train_loader = DataLoader(train_ds, batch_size=BATCH, shuffle=True,  num_workers=4, pin_memory=True)
val_loader   = DataLoader(val_ds,   batch_size=BATCH, shuffle=False, num_workers=4, pin_memory=True)
test_loader  = DataLoader(test_ds,  batch_size=BATCH, shuffle=False, num_workers=4, pin_memory=True)

len(train_ds), len(val_ds), len(test_ds), train_ds.classes
```

```python
import torch.nn as nn
from torchvision import models

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = models.resnet18(weights=None)
model.fc = nn.Linear(model.fc.in_features, len(SHAPES))
model.to(device)

crit = nn.CrossEntropyLoss()
opt  = torch.optim.AdamW(model.parameters(), 1e-3)

for epoch in range(5):
    model.train()
    tot=0
    for x,y in train_loader:
        x,y = x.to(device), y.to(device)
        opt.zero_grad()
        out = model(x)
        loss = crit(out,y)
        loss.backward()
        opt.step()
        tot += loss.item()*x.size(0)
    tr_loss = tot/len(train_ds)

    model.eval()
    correct=0
    with torch.no_grad():
        for x,y in val_loader:
            x,y = x.to(device), y.to(device)
            pred = model(x).argmax(1)
            correct += (pred==y).sum().item()
    val_acc = correct/len(val_ds)
    print(f"Epoch {epoch+1}: loss={tr_loss:.4f} val_acc={val_acc:.3f}")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import AdamW
from torch.cuda.amp import autocast, GradScaler
from torch.optim.lr_scheduler import CosineAnnealingLR
import copy, math, time

# ---------------- Model ----------------
class DWConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch, stride=1):
        super().__init__()
        self.dw = nn.Conv2d(in_ch, in_ch, 3, stride=stride, padding=1, groups=in_ch, bias=False)
        self.pw = nn.Conv2d(in_ch, out_ch, 1, bias=False)
        self.bn = nn.BatchNorm2d(out_ch)
        self.act = nn.ReLU(inplace=True)
    def forward(self, x):
        x = self.dw(x)
        x = self.pw(x)
        x = self.bn(x)
        return self.act(x)

class TinyConvNet(nn.Module):
    def __init__(self, num_classes=3):
        super().__init__()
        self.stem = nn.Sequential(
            nn.Conv2d(3, 32, 3, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True)
        )
        self.block1 = DWConvBlock(32, 64, stride=1)
        self.block2 = DWConvBlock(64, 128, stride=2)
        self.block3 = DWConvBlock(128, 128, stride=1)
        self.block4 = DWConvBlock(128, 256, stride=2)
        self.head = nn.Sequential(
            nn.Conv2d(256, 256, 1, bias=False),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True)
        )
        self.classifier = nn.Linear(256, num_classes)

    def forward(self, x):
        x = self.stem(x)
        x = self.block1(x)
        x = self.block2(x)
        x = self.block3(x)
        x = self.block4(x)
        x = self.head(x)
        x = F.adaptive_avg_pool2d(x, 1).flatten(1)   # <- FIX HERE
        return self.classifier(x)

# --------------- Train loop ---------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def train_model(model, train_loader, val_loader,
                epochs=15, lr=3e-4, weight_decay=1e-4,
                patience=5, ckpt_path="best_model.pt",
                mixed_precision=True):

    model = model.to(device)
    crit = nn.CrossEntropyLoss()
    opt = AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)
    sched = CosineAnnealingLR(opt, T_max=epochs, eta_min=1e-6)
    scaler = GradScaler(enabled=(mixed_precision and device.type=='cuda'))

    best_weights = copy.deepcopy(model.state_dict())
    best_val_loss = math.inf
    no_improve = 0

    for epoch in range(1, epochs+1):
        t0 = time.time()
        model.train()
        train_loss = 0.0
        for x,y in train_loader:
            x,y = x.to(device, non_blocking=True), y.to(device, non_blocking=True)
            opt.zero_grad(set_to_none=True)
            with autocast(enabled=(mixed_precision and device.type=='cuda')):
                out = model(x)
                loss = crit(out, y)
            scaler.scale(loss).backward()
            scaler.step(opt)
            scaler.update()
            train_loss += loss.item()*x.size(0)
        train_loss /= len(train_loader.dataset)

        model.eval()
        val_loss, correct = 0.0, 0
        with torch.no_grad():
            for x,y in val_loader:
                x,y = x.to(device), y.to(device)
                out = model(x)
                loss = crit(out, y)
                val_loss += loss.item()*x.size(0)
                correct += (out.argmax(1)==y).sum().item()
        val_loss /= len(val_loader.dataset)
        val_acc = correct/len(val_loader.dataset)

        sched.step()
        elapsed = time.time() - t0
        print(f"Epoch {epoch:02d}/{epochs} | {elapsed:.1f}s | "
              f"train_loss={train_loss:.4f} | val_loss={val_loss:.4f} | val_acc={val_acc:.3f}")

        if val_loss < best_val_loss - 1e-6:
            best_val_loss = val_loss
            best_weights = copy.deepcopy(model.state_dict())
            torch.save(best_weights, ckpt_path)
            no_improve = 0
        else:
            no_improve += 1
            if no_improve >= patience:
                print("Early stopping triggered.")
                break

    model.load_state_dict(best_weights)
    return model

# --------------- Run ---------------
num_classes = 3
model = TinyConvNet(num_classes)
model = train_model(model, train_loader, val_loader,
                    epochs=15, lr=3e-4, weight_decay=1e-4,
                    patience=5, ckpt_path="best_tinyconv.pt")
```

```python
def train_model(model, train_loader, val_loader,
                epochs=15, lr=3e-4, weight_decay=1e-4,
                patience=5, ckpt_path="best_model.pt",
                mixed_precision=True):

    model = model.to(device)
    crit = nn.CrossEntropyLoss()
    opt = AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)
    sched = CosineAnnealingLR(opt, T_max=epochs, eta_min=1e-6)
    scaler = GradScaler(enabled=(mixed_precision and device.type=='cuda'))

    best_weights = copy.deepcopy(model.state_dict())
    best_val_loss = math.inf
    no_improve = 0

    history = {"train_loss": [], "val_loss": [], "val_acc": []}

    for epoch in range(1, epochs+1):
        t0 = time.time()
        model.train()
        train_loss = 0.0
        for x,y in train_loader:
            x,y = x.to(device, non_blocking=True), y.to(device, non_blocking=True)
            opt.zero_grad(set_to_none=True)
            with autocast(enabled=(mixed_precision and device.type=='cuda')):
                out = model(x)
                loss = crit(out, y)
            scaler.scale(loss).backward()
            scaler.step(opt)
            scaler.update()
            train_loss += loss.item()*x.size(0)
        train_loss /= len(train_loader.dataset)

        model.eval()
        val_loss, correct = 0.0, 0
        with torch.no_grad():
            for x,y in val_loader:
                x,y = x.to(device), y.to(device)
                out = model(x)
                loss = crit(out, y)
                val_loss += loss.item()*x.size(0)
                correct += (out.argmax(1)==y).sum().item()
        val_loss /= len(val_loader.dataset)
        val_acc = correct/len(val_loader.dataset)

        sched.step()
        elapsed = time.time() - t0
        print(f"Epoch {epoch:02d}/{epochs} | {elapsed:.1f}s | "
              f"train_loss={train_loss:.4f} | val_loss={val_loss:.4f} | val_acc={val_acc:.3f}")

        history["train_loss"].append(train_loss)
        history["val_loss"].append(val_loss)
        history["val_acc"].append(val_acc)

        if val_loss < best_val_loss - 1e-6:
            best_val_loss = val_loss
            best_weights = copy.deepcopy(model.state_dict())
            torch.save(best_weights, ckpt_path)
            no_improve = 0
        else:
            no_improve += 1
            if no_improve >= patience:
                print("Early stopping triggered.")
                break

    model.load_state_dict(best_weights)
    return model, history
```

```python
model = TinyConvNet(num_classes=3)
model, history = train_model(model, train_loader, val_loader,
                             epochs=30, lr=3e-4, weight_decay=1e-4,
                             patience=7, ckpt_path="best_tinyconv.pt")
```
